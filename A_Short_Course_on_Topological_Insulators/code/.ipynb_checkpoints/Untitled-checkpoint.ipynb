```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as lin
```

```python
def ham_ssh(v,w,kl):
    vals = []
    for k in kl:
        H = np.array([[0, v+w*np.exp(-1j*k)],[v+w*np.exp(1j*k), 0]])
        val = lin.eigvalsh(H)
        vals.append(val)
    return vals

def dxdy(v,w,kl):
    return (v+w*np.cos(kl),w*np.sin(kl))
```

```python
klist=np.linspace(-np.pi,np.pi)

fig = plt.figure(figsize=[18,12]);
ax1 = plt.subplot(251)
ax1.set_box_aspect(1)
ax1.plot(klist, ham_ssh(1,0,klist))

ax2 = plt.subplot(252)
ax2.set_box_aspect(1)
ax2.plot(klist, ham_ssh(1,1/2,klist))

ax3 = plt.subplot(253)
ax3.set_box_aspect(1)
ax3.plot(klist, ham_ssh(1,1,klist))

ax4 = plt.subplot(254)
ax4.set_box_aspect(1)
ax4.plot(klist, ham_ssh(1/2,1,klist))

ax5 = plt.subplot(255)
ax5.set_box_aspect(1)
ax5.plot(klist, ham_ssh(0,1,klist))

ax6 = plt.subplot(256)
ax6.set_box_aspect(1)
ax6.plot(dxdy(1,0,klist)[0],dxdy(1,0,klist)[1], marker='d')

ax7 = plt.subplot(257)
ax7.set_box_aspect(1)
ax7.plot(dxdy(1,1/2,klist)[0],dxdy(1,1/2,klist)[1], marker='d')

ax8 = plt.subplot(258)
ax8.set_box_aspect(1)
ax8.plot(dxdy(1,1,klist)[0],dxdy(1,1,klist)[1], marker='d')

ax9 = plt.subplot(259)
ax9.set_box_aspect(1)
ax9.plot(dxdy(1/2,1,klist)[0],dxdy(1/2,1,klist)[1], marker='d')

ax0 = plt.subplot(2,5,10)
ax0.set_box_aspect(1)
ax0.plot(dxdy(0,1,klist)[0],dxdy(0,1,klist)[1], marker='d')
```
